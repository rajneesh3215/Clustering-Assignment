# K-Means Clustering

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
airline1.info()
```

```python
airline1.isnull().sum()
```

```python
#Standardize the data
airline_norm=StandardScaler().fit_transform(airline1)
```

```python
airline_norm
```

```python
# By using WCSS criteria find k cluster
wcss=[]
for i in range(1,12):
    kmeans=KMeans(n_clusters=i,random_state=0)
    kmeans.fit(airline_norm)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,12),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of cluster')
plt.ylabel('wcss')
plt.show()
```

```python
cluster=KMeans(6,random_state=42)
cluster.fit(airline_norm)
```

```python
cluster.labels_
```

```python
airline1['cluster']=cluster.labels_
```

```python
airline1.groupby('cluster').agg(['mean']).reset_index()
```
